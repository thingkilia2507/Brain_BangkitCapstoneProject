```python
dataset = pd.read_csv('joy-ratings-0to1.test.gold.txt', delimiter='\t', header=None)
```

```python
used = dataset[500:600].copy()
```

```python
translator = Translator()
```

```python
def translate(conversation):
    dialogue_translated = translator.translate(conversation[1], dest='id').text
    return pd.Series([dialogue_translated])
```

```python
test = used.copy()
test['dialogueTranslated'] = test.apply(lambda conversation: translate(conversation), axis=1)
test
```

```python
with open('joy-test-translated-5.csv', 'w') as f:
    test.to_csv(f, header=False, index=False)
```
